## 2. Verify GPU access

PyTorch should report **`True`** for CUDA if you plan to train on GPU.

- If you see **`False`**, training falls back to CPU and will be very slow.
- **Fix:** install a CUDA-enabled PyTorch build, or use a cloud GPU (e.g. Colab with GPU runtime).

```python
import torch

print(torch.cuda.is_available())  # should be True on a CUDA machine
if torch.cuda.is_available():
    print(torch.cuda.get_device_name(0))  # e.g. "Tesla T4"
```
